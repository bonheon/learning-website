```python
plt.plot("Index","Y",data=vis_data,
        marker="X",
        markersize=15,
        markerfacecolor="pink",
        markeredgecolor="green")
plt.show()

##https://matplotlib.org/stable/api/markers_api.html##
```

```python
plt.plot("Index","Y",data=vis_data,
         linewidth=4,
        linestyle="dotted",
        color="orange")
plt.show()

##https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
```

```python
plt.plot("Index","Y",data=vis_data,
        linewidth=10,
        linestyle="dashed",
        color="orange",
        marker="o",
        markersize=10,
        markerfacecolor="pink",
        markeredgecolor="green")
plt.show()
```

```python
plt.plot("Index","Y",data=vis_data,
        linewidth=10,
        linestyle="dashed",
        color="orange",
        marker="o",
        markersize=10,
        markerfacecolor="pink",
        markeredgecolor="green")
plt.axvline(x=5,color="black",linewidth=3)
plt.axhline(y=np.mean(vis_data["Y"]),color="red",linewidth=3)
plt.show()
```
