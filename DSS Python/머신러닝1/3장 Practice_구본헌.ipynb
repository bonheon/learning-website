# Practice 1

```python
import pandas as pd
import matplotlib.pyplot as plt   # 데이터 시각화 라이브러리 및 모듈
import seaborn as sns             # 데이터 시각화 라이브러리

plt.rcParams["axes.unicode_minus"]=False
plt.rcParams["font.family"]="AppleGothic"


prac_data=pd.read_csv("data/Preprocessing_Data.csv",encoding="EUC-KR")


prac_data
```

```python
#1) 'x3'의 Range 및 Interquartile Range(IQR)를 구하시오 (Range : 최대값 - 최소값)

x3_min = prac_data['x3'].min()
x3_max = prac_data['x3'].max()
x3_q1 = prac_data['x3'].quantile(0.25)
x3_q3 = prac_data['x3'].quantile(0.75)

x3_range = x3_max - x3_min
x3_IQR = x3_q3 - x3_q1

print('x3의 Range : ', x3_range)
print('x3의 IQR : ', x3_IQR)
```

```python
#2) 결측치가 존재하는 변수들을 모두 찾아보시오.

print('결측치가 존재하는 변수들은 : ', prac_data.columns[prac_data.isnull().sum() > 0].values)
```

```python
#3) Seaborn 또는 matplotlib 라이브러리를 사용하여 'x4'와 'y'의 산점도(Scatterplot)를 그려보고, y의 평균값과 
#   x4의 평균값을 갖는 수직성과 수평선을 그려보시오.

sns.scatterplot(data=prac_data, x='x4', y='y')
plt.axhline(y=prac_data['y'].mean(), color='black', linewidth=1)
plt.axvline(x=prac_data['x4'].mean(), color='black', linewidth=1)
```

```python
#(a) 입력변수 중 고유값 개수가 3미만인 변수를 제거한 데이터셋을 EQ_data2에 저장하고, EQ_data2의 크기를 출력하시오.
# 몇 개의 변수가 줄었는가?

EQ_data2 = EQ_data.loc[:,EQ_data.columns[~(EQ_data.nunique() <= 3)]]
```

```python
print('데이터가 {0}개 줄었다.'.format(EQ_data.shape[1] - EQ_data2.shape[1]))
```

```python
#(b) EQ_data2에서 결측치 수가 전체 데이터의 5% 이상인 입력변수를 제거하시오.
#  몇 개의 변수가 줄었는가?
print('기존 EQ_data2의 변수의 수 : ', EQ_data2.shape[1])
drop_col = EQ_data2.columns[EQ_data2.isnull().sum() > EQ_data2.shape[0] * 0.05]
```

```python
EQ_data2.drop(columns=drop_col, inplace=True)
print('제거 후 변수의 수 ', EQ_data2.shape[1]) #2개 감소하였음
```

```python
#(c) EQ_data2에서 결측치 수가 3개 이상인 행을 모두 제거하시오. 몇 개의 데이터가 줄었는가? 
print('기존 EQ_data2의 샘플 수 : ', EQ_data2.shape[0])
drop_idx = EQ_data2.index[EQ_data2.isnull().sum(axis=1) >= 3]
EQ_data2.drop(index=drop_idx, inplace=True)
print('제거 후 샘플의 수 ', EQ_data2.shape[0]) # 74개 감소하였음
```

```python
#(d) 결측치가 1개 이상인 변수명을 모두 출력하시오.
EQ_data2.columns[EQ_data2.isnull().sum() >= 1]
```

```python
#(e) 연속형 변수의 결측값은 각 변수릐 평균값으로 대치하시오.
EQ_data2.fillna(EQ_data2.mean(numeric_only=True), inplace=True)
```

```python
#(f) 범주형 변수의 결측값은 'Unknown'으로 대치하시오.
EQ_data2.fillna('Unknown', inplace=True)
```

```python
#(g) 'A3'변수를 기준으로 평균 +_6 표준편차를 벗어나는 이상치를 탐지하시오. 총 몇개의 데이터가 탐지되었는가?

outlier_data=EQ_data2.loc[((EQ_data2['A3']>EQ_data2['A3'].mean()+EQ_data2['A3'].std()*6)|
                           (EQ_data2['A3']<EQ_data2['A3'].mean()-EQ_data2['A3'].std()*6)),:]
```

```python
print('총 {0}개의 데이터가 탐지되었음'.format(outlier_data.shape[0]))
```

```python
#(h) EQ_data2에서 원-핫 인코딩을 수행한 데이터를 EQ_data3에 저장하시오. 이 때 EQ_data3의 변수의 수는 몇 개 인가?

EQ_data3 = pd.get_dummies(EQ_data2)
print('EQ_data3의 변수의 수 : ', EQ_data3.shape[1])
```
