# 1. Bar Plot

- Bar Plot은 이산형 자료에 대해서 범주(Category), 계급(Class)별로 빈도나 합계 등을 막대로 그려서 비교하는데 유용하게 사용하는 막대그래프

## 1-1. 요일별 Tip의 합계

```python
# 데이터를 그룹화하고 요일별 팁 합계를 구함
tip_day = tips.groupby('day').tip.sum()
tip_day
```

```python
# tip_day에 있는 요일을 list로 추출
days = tip_day.index.to_list()
days
```

```python
# Bar plot 그리기 (seaborn)
# 그래프의 크기 설정
plt.figure(figsize=(10, 8))

# Seaborn을 사용한 수직 막대 그래프(bar plot) 그리기
# x: x 축에 표시할 열 (요일, days)
# y: y 축에 표시할 열 (팁 합계, tip_day)
# palette: 색상 팔레트 설정
sns.barplot(x=days, y=tip_day, palette='Set2')

# x 축 눈금 레이블 설정 (폰트 크기와 눈금 라벨 회전)
plt.xticks(fontsize=15, rotation=30)

# y 축 눈금 레이블 설정 (폰트 크기)
plt.yticks(fontsize=15)

# 그래프 제목 설정 (폰트 크기)
plt.title('Sum of Tips by Day', fontsize=20)

# x 축 레이블 설정 (폰트 크기)
plt.xlabel('Day', fontsize=18)

# y 축 레이블 설정 (폰트 크기)
plt.ylabel('Sum of Tips', fontsize=18)

# 그래프 표시
plt.show()
```

```python
# Bar plot 그리기 (matplotlib)
# 그래프의 크기 설정
plt.figure(figsize=(10, 8))

# x 축과 y 축 데이터를 이용하여 막대 그래프(bar plot) 그리기
plt.bar(days, tip_day)

# x 축 눈금 레이블 설정 (폰트 크기와 눈금 라벨 회전)
plt.xticks(fontsize=15, rotation=30)

# y 축 눈금 레이블 설정 (폰트 크기)
plt.yticks(fontsize=15)

# 그래프 제목 설정 (폰트 크기)
plt.title('Sum of Tips by Day', fontsize=20)

# x 축 레이블 설정 (폰트 크기)
plt.xlabel('Day', fontsize=18)

# y 축 레이블 설정 (폰트 크기)
plt.ylabel('Sum of Tips', fontsize=18)

# 그래프 표시
plt.show()
```

# 2. Pie Plot

- 파이 차트 (Pie chart, 원 그래프)는 범주별 구성 비율을 원형으로 표현한 그래프
- 부채꼴의 중심각을 구성 비율에 비례하도록 표현

```python
# 각 요일별 팁의 합계 값을 가져옴
values = tip_day.values
values

# 요일 이름을 라벨로 사용
labels = tip_day.index
labels
```

```python
# 파이 차트 그리기
plt.figure(figsize=(10, 8))  # 그림의 크기 설정

# values: 데이터 값, labels: 라벨
plt.pie(values, labels=labels)

# 그래프 제목 추가
plt.title('Sum of Tips by Day - Pie Plot', fontsize=20)

# 범례 추가 (라벨을 요일 이름으로 설정)
plt.legend(labels)

# 그래프 표시
plt.show()
```

```python
# 파이 차트 스타일 및 설정
explode = [0, 0.15, 0, 0]  # 부채꼴이 떨어져 보이도록 설정
startangle = 90  # 시작 각도를 90도로 설정 (12시 방향에서 시작)
colors = ['yellow', 'skyblue', 'purple', 'pink']  # 부채꼴 색상 지정

plt.figure(figsize=(10, 8))  # 그림의 크기 설정

# values: 데이터 값, labels: 라벨, explode: 부채꼴 튀어나오는 정도, startangle: 시작 각도, colors: 색상
plt.pie(values, labels=labels, explode=explode, startangle=startangle, colors=colors)

# 그래프 제목 추가
plt.title('Sum of Tips by Day - Pie Plot', fontsize=20)

# 범례 추가 (라벨을 요일 이름으로 설정)
plt.legend(labels)

# 그래프 표시
plt.show()
```

```python
plt.figure(figsize=(10, 8))  # 그림의 크기 설정

# values: 데이터 값, labels: 라벨
# autopct: 각 부채꼴에 백분율 값 추가
plt.pie(values, labels=labels, autopct='%.1f%%')

# 그래프 제목 추가
plt.title('Sum of Tips by Day - Pie Plot', fontsize=20)

# 범례 추가 (라벨을 요일 이름으로 설정)
plt.legend(labels)

# 그래프 표시
plt.show()
```

# 5. Box Plot

- Box Plot은 자료에서 얻은 다섯가지의 통계량을 기반으로 그린 그래프임
- 최솟값 : 제 1사분위 - 1.5 * IQR
- 제 1사분위(Q1) : 25% 위치
- 제 2사분위(Q2) : 50% 위치, 중앙값(median)
- 제 3사분위(Q3) : 75% 위치
- 최댓값 : 제 3사분위 + 1.5 * IQR
- Q1 - 1.5 * IQR 보다 작거나 Q3 - 1.5 * IQR 보다 크면 이상치로 판단

```python
# 그래프의 크기 설정
plt.figure(figsize=(10, 8))

# 그래프 제목 설정
plt.title('Total bill by Time')

# Seaborn을 사용한 상자 그림(boxplot) 그리기
# x: x 축에 표시할 열 (시간)
# y: y 축에 표시할 열 (총 금액)
# data: 사용할 데이터프레임 (tips 데이터 사용)
sns.boxplot(x='time', y='total_bill', data=tips)

# 그래프 표시
plt.show()
```

```python
# 그래프의 크기 설정
plt.figure(figsize=(10, 8))

# 그래프 제목 설정
plt.title('Total bill by Day')

# Seaborn을 사용한 상자 그림(boxplot) 그리기
# x: x 축에 표시할 열 (요일)
# y: y 축에 표시할 열 (총 금액)
# hue: 시간에 따라 색상을 다르게 표시
# data: 사용할 데이터프레임 (tips 데이터 사용)
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)

# 범례 추가
plt.legend(title='Time')

# 그래프 표시
plt.show()
```

# 12. Q-Q Plot

- 확률 분포나 데이터의 분포를 시각적으로 평가하고, 정규 분포 또는 다른 특정 분포와의 일치를 확인하는 데 사용되는 그래프
- Q-Q plot은 데이터의 분포를 정규 분포와 비교하거나, 다른 분포와 비교하여 데이터의 정규성 검사, 이상치 탐지 및 분포 적합성 검사에 유용함

```python
from scipy.stats import probplot

# CSV 파일에서 데이터 불러오기
df_normal = pd.read_csv(path + 'normal_data.csv')
df_heavy_tailed = pd.read_csv(path + 'heavy_tailed_data.csv')
df_left_skewed = pd.read_csv(path + 'left_skewed_data.csv')

# Q-Q Plot 그리기
f, axes = plt.subplots(1, 3, figsize=(15, 5))

# 정규 분포 데이터의 Q-Q Plot
probplot(df_normal['Normal'], plot=axes[0])
axes[0].set_title('Q-Q Plot of Normal Data')
axes[0].set_xlabel('Theoretical Quantiles')
axes[0].set_ylabel('Ordered Values')

# Heavy-tailed 분포 데이터의 Q-Q Plot
probplot(df_heavy_tailed['Heavy-Tailed'], plot=axes[1])
axes[1].set_title('Q-Q Plot of Heavy-Tailed Data')
axes[1].set_xlabel('Theoretical Quantiles')
axes[1].set_ylabel('Ordered Values')

# Left-skewed 분포 데이터의 Q-Q Plot
probplot(df_left_skewed['Left-Skewed'], plot=axes[2])
axes[2].set_title('Q-Q Plot of Left-Skewed Data')
axes[2].set_xlabel('Theoretical Quantiles')
axes[2].set_ylabel('Ordered Values')

# 그래프 표시
plt.tight_layout()
plt.show()
```
