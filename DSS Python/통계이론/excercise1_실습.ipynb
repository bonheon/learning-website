# Matplotlib vs Seaborn
- Matplotlib는 Python 프로그래밍 언어 및 수학적 확장 NumPy 라이브러리를 활용한 플로팅 라이브러리

- seaborn은 matplotlib 를 기반으로 하는 Python 데이터 시각화 라이브러리이며, 매력적이고 유익한 통계 그래픽을 그리기 위한 고급 인터페이스를 제공

# 5. Box Plot

- Box Plot은 자료에서 얻은 다섯가지의 통계량을 기반으로 그린 그래프임
- 최솟값 : 제 1사분위 - 1.5 * IQR
- 제 1사분위(Q1) : 25% 위치
- 제 2사분위(Q2) : 50% 위치, 중앙값(median)
- 제 3사분위(Q3) : 75% 위치
- 최댓값 : 제 3사분위 + 1.5 * IQR
- Q1 - 1.5 * IQR 보다 작거나 Q3 - 1.5 * IQR 보다 크면 이상치로 판단

# 12. Q-Q Plot

- 확률 분포나 데이터의 분포를 시각적으로 평가하고, 정규 분포 또는 다른 특정 분포와의 일치를 확인하는 데 사용되는 그래프
- Q-Q plot은 데이터의 분포를 정규 분포와 비교하거나, 다른 분포와 비교하여 데이터의 정규성 검사, 이상치 탐지 및 분포 적합성 검사에 유용함

```python
from scipy.stats import probplot

# CSV 파일에서 데이터 불러오기
df_normal = pd.read_csv(path + 'normal_data.csv')
df_heavy_tailed = pd.read_csv(path + 'heavy_tailed_data.csv')
df_left_skewed = pd.read_csv(path + 'left_skewed_data.csv')

# Q-Q Plot 그리기
f, axes = plt.subplots(1, 3, figsize=(15, 5))

# 정규 분포 데이터의 Q-Q Plot
probplot(df_normal['Normal'], plot=axes[0])
axes[0].set_title('Q-Q Plot of Normal Data')
axes[0].set_xlabel('Theoretical Quantiles')
axes[0].set_ylabel('Ordered Values')

# Heavy-tailed 분포 데이터의 Q-Q Plot
probplot(df_heavy_tailed['Heavy-Tailed'], plot=axes[1])
axes[1].set_title('Q-Q Plot of Heavy-Tailed Data')
axes[1].set_xlabel('Theoretical Quantiles')
axes[1].set_ylabel('Ordered Values')

# Left-skewed 분포 데이터의 Q-Q Plot
probplot(df_left_skewed['Left-Skewed'], plot=axes[2])
axes[2].set_title('Q-Q Plot of Left-Skewed Data')
axes[2].set_xlabel('Theoretical Quantiles')
axes[2].set_ylabel('Ordered Values')

# 그래프 표시
plt.tight_layout()
plt.show()
```
