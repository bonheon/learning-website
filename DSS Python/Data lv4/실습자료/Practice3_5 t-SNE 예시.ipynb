```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from mpl_toolkits.mplot3d import axes3d
```

```python
X = pd.read_csv("data/X_t_sne.csv")
y = pd.read_csv("data/y_t_sne.csv")
```

```python
temp = X.values
```

```python
y = y.values.reshape(-1)
```

```python
cluster0 = temp[y==0]
cluster1 = temp[y==1]
```

```python
%matplotlib notebook
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(cluster0[:, 0], cluster0[:, 1], cluster0[:, 2], marker='o', c='blue', s=20)
ax.scatter(cluster1[:, 0], cluster1[:, 1], cluster1[:, 2], marker='o', c='orange', s=20)
plt.show()
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(perplexity=2)
```

```python
embeddings = tsne.fit_transform(temp)
```

```python
cluster0 = embeddings[y==0]
cluster1 = embeddings[y==1]
```

```python
fig = plt.figure(figsize=(5, 5))
plt.scatter(cluster0[:, 0], cluster0[:, 1], marker='o', c='blue', s=20)
plt.scatter(cluster1[:, 0], cluster1[:, 1], marker='o', c='orange', s=20)
plt.show()
```
