- num_parts: 연속형 입력변수의 분할 수
- cset_max_length: 생성된 cset의 최대 길이

- min_group_sup: 허용 가능한 특정 그룹 내 cset의 minimum support 
- min_total_sup: 허용 가능한 전체 데이터 내 cset의 minimum support 
- max_total_sup: 허용 가능한 전체 데이터 내 cset의 maximum support

- alpha: 독립성 검정을 위한 유의수준 기준
- min_dev: 허용 가능한 그룹 간 cset의 support 차이 
- min_corr: 허용 가능한 cset과 출력변수와의 minimum correlation

- min_subset_sup: 허용 가능한 부분 집합과의 최소 support 차이 
- min_subset_corr: 허용 가능한 부분 집합과의 최소 correlation 차이

- Group support는 우리가 관심있는 대상인 Fault (Group)에 대해 찾아낸 규칙이 얼마나 Fault를 많이 포함하고 있는 지
- Total Support는 찾아낸 규칙이 전체 데이터 중 얼마나 많이 포함하고 있는 지
- Lift는 찾아낸 규칙, 관심있는 대상을 각각 A, B로 봤을 때 P(A∩B)/P(A)P(B)로 B 대비 A와 B인 경우로 1보다 크면 우수한 규칙임을 의미

```python
plt.scatter(y[(data["x3"]=="B3") & (data["x7"]=="A7")].index.tolist(), y[(data["x3"]=="B3") & (data["x7"]=="A7")], color="red", label='x3_B3 & x7_A7', s=5)
plt.scatter(y[~((data["x3"]=="B3") & (data["x7"]=="A7"))].index.tolist(), y[~((data["x3"]=="B3") & (data["x7"]=="A7"))], color="dodgerblue", label='Others', s=5)
plt.hlines(50, 0, 1000, color='red')
plt.legend()
plt.show()
```
